```python
import pandas as pd
```

```python
## Match Results Data
```

```python
match_result=pd.read_json(r'D:\cricket Analysis using PowerBI\t20_json_files\t20_wc_match_results.json')
```

```python
match_result
```

```python
match_result_data=pd.json_normalize(match_result['matchSummary'].explode())
```

```python
match_result_data.head()
```

```python
match_result_data.shape
```

```python
match_result_data.rename(columns={'scorecard':'match_id'},inplace=True)
```

```python
match_result_data.head()
```

```python
match_ids_dict={}
for index,row in match_result_data.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
match_ids_dict
```

```python
match_result_data.to_csv('matchsummary.csv',index=False)
```
